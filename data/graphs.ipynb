```python
import pandas as pd
import plotly.express as px

# Load the data
data = pd.read_csv('words_with_sm_norms.csv')

# Define colors for the 11 sensorimotor categories
colors = {
    "haptic": "#ffdf00",
    "auditory": "#c71585",
    "olfactory": "#00ff00",
    "gustatory": "#00bfff",
    "visual": "#ff4500",
    "interoceptive": "#7c550c",
}

new_labels = {
    "haptic": "Touch",
    "auditory": "Sound",
    "olfactory": "Smell",
    "gustatory": "Taste",
    "visual": "Sight",
    "interoceptive": "Interoception",
}

# Create a new column 'color' in the DataFrame that contains the color for each row
data['color'] = data['Dominant.perceptual'].map(colors)
data['new_labels'] = data['Dominant.perceptual'].map(new_labels)

df = px.data.tips()
fig = px.pie(data, names='new_labels', color='Dominant.perceptual', color_discrete_map=colors,
             labels={
                    'new_labels': 'Sensory category',
                    'color': 'Sensory category'
             })
fig.update_layout(width=600, height=500, font_family="Garamond", font_size=10)
fig.for_each_trace(lambda trace: trace.update(name=new_labels[trace.name] if trace.name in new_labels else trace.name))
fig.show()
fig.write_image("figures/total_sense_pie.svg")
# Filter the data based on the 'appears_in' column
data_portraits_titles = data[data['appears_in'].str.contains('po|to')]
data_only_portraits = data[data['appears_in'].str.contains('po')]
data_only_titles = data[data['appears_in'].str.contains('to')]

# Extract the numerical position from the 'appears_in' column
data_portraits_titles['position'] = data_portraits_titles['appears_in'].str.extract('(\d+)').astype(int)
data_only_portraits['position'] = data_only_portraits['appears_in'].str.extract('(\d+)').astype(int)
data_only_titles['position'] = data_only_titles['appears_in'].str.extract('(\d+)').astype(int)

# Define the ranges for the three subsections
ranges = {
    'Objects': (0, 57),
    'Food': (58, 108),
    'Rooms': (109, 109)
}

# Create the variant pie charts for each subsection
for name, (start, end) in ranges.items():
    for data_subset in [data_portraits_titles, data_only_portraits, data_only_titles]:
        # Filter the data based on the numerical position
        data_filtered = data_subset[(data_subset['position'] >= start) & (data_subset['position'] <= end)]
        
        # Create the pie chart
        fig = px.pie(data_filtered, names='new_labels', color='Dominant.perceptual', color_discrete_map=colors,
                     labels={
                            'new_labels': 'Sensory category',
                            'color': 'Sensory category'
                     })
        fig.update_layout(width=600, height=500, font_family="Garamond", font_size=10)
        fig.for_each_trace(lambda trace: trace.update(name=new_labels[trace.name] if trace.name in new_labels else trace.name))
        fig.show()
        fig.write_image(f"figures/{name}_{'all' if data_subset is data_portraits_titles else 'portraits' if data_subset is data_only_portraits else 'titles'}.svg")
```

```python
fig = px.box(data, x='Dominant.perceptual', y='Exclusivity.perceptual', color='Dominant.perceptual', color_discrete_map=colors,
                 labels={
                        'Dominant.perceptual': 'Dominant sensory category',
                        'Exclusivity.perceptual': 'Exclusivity of sensory category'
                 })
fig.update_layout(width=800, height=500, font_family="Garamond", font_size=10, showlegend=False,
                  plot_bgcolor='rgb(243, 243, 243)')
fig.for_each_trace(lambda trace: trace.update(name=new_labels[trace.name] if trace.name in new_labels else trace.name))
fig.update_traces(boxpoints='all', jitter=0.4, whiskerwidth=0.2, marker_size=2, line_width=1)
fig.update_traces(line_color='black', marker_line_width=0.2, marker_line_color='black', marker_opacity=0.4)
# Define the new labels
new_x_labels = {
    "haptic": "Touch",
    "auditory": "Sound",
    "olfactory": "Smell",
    "gustatory": "Taste",
    "visual": "Sight",
    "interoceptive": "Interoception",
}

# Update the x-axis labels
fig.update_xaxes(
    ticktext=list(new_x_labels.values()),
    tickvals=list(new_x_labels.keys())
)

fig.show()
fig.write_image("figures/exclusivity_boxplot.svg")
```
